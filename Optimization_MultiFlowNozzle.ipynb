```python

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df  = pd.read_csv("/content/Nozzle-data.csv")
```

```python
df
```

```python
x_values = df[["Air Hole Dia",	"Ball Dia"]].values
x_1 = df["Air Hole Dia"].values
x_2 = df["Ball Dia"].values
y_values1 = df["Display height"].values
y_values2 = df["Vol frac %"].values
```

```python
230/3700
```

```python
x_2
```

```python
print(x_values[0], y_values2[0])
```

```python
degree = 3
poly_model = PolynomialFeatures(degree=degree)
poly_x_values = poly_model.fit_transform(x_values)
print(f'initial values {x_values[0]}\nMapped to {poly_x_values[0]}')
```

```python
poly_model.fit(poly_x_values, y_values1)
regression_model = LinearRegression()

regression_model.fit(poly_x_values, y_values1)

y_pred = regression_model.predict(poly_x_values)

regression_model.coef_

mean_squared_error(y_values1, y_pred, squared=False)
```

```python
fig, axs = plt.subplots(1, figsize=(15,15))
ax = plt.axes(projection ="3d")

ax.scatter3D(x_1,x_2, y_values1, color = "green")
ax.scatter3D(x_1,x_2, y_pred, color = "red")
plt.title("simple 3D scatter plot")
ax.set_xlabel("Air Hole Dia")
ax.set_ylabel("Ball dia")
ax.set_zlabel("Display height",fontsize = 10)
```

```python
poly_model.fit(poly_x_values, y_values2)
regression_model2 = LinearRegression()

regression_model2.fit(poly_x_values, y_values2)

y_pred2 = regression_model2.predict(poly_x_values)

regression_model2.coef_

mean_squared_error(y_values2, y_pred2, squared=False)
```
